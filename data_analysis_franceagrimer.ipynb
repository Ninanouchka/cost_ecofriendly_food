```python
df_full = df_full.dropna(subset=['MARCHE'])
```

```python
df_full['label'] = df_full['MARCHE'].apply(lambda x: 1 if x.lower().find('bio') > -1 else 0)
df_full['origin'] = df_full['STADE'].apply(lambda x: 1 if x.lower().find('import') > -1 else 0)
print(df_full.shape)
# df_full = df_full.drop(df_full[df_full['STADE'] == 'Exp�dition'].index)
print(df_full.shape)
df_full = df_full.drop(df_full[df_full['UNITE PRIX'].str.find('le kg') == -1].index)
print(df_full.shape)
```

```python
df_data = df_full[['LIBELLE', 'STADE', 'MOY', 'label', 'origin']]
```

#### Calcul price organic vs not organic

```python
df_data['prix_bio'] = df_data.apply(lambda x: float(x['MOY']) if x['label'] == 1 else np.nan, axis=1)
df_data['prix_norm'] = df_data.apply(lambda x: float(x['MOY']) if x['label'] == 0 else np.nan, axis=1)
```

```python
plt.boxplot([df_data['prix_bio'].dropna().values, df_data['prix_norm'].dropna().values], showfliers=False)
plt.xticks([1, 2], ['Organic', 'Not organic'])
plt.ylabel('Price (euro)')
plt.title('Price  per kg')
```

#### Calcul price imported vs local

```python
df_data['prix_import'] = df_data.apply(lambda x: float(x['MOY']) if x['origin'] == 1 else np.nan, axis=1)
df_data['prix_interieur'] = df_data.apply(lambda x: float(x['MOY']) if x['origin'] == 0 else np.nan, axis=1)
```

```python
plt.boxplot([df_data['prix_import'].dropna().values, df_data['prix_interieur'].dropna().values], showfliers=False)
plt.xticks([1, 2], ['Imported', 'France'])
plt.ylabel('Price (euro)')
plt.title('Price of fresh fruits per kg')
```

```python
def dist_origin(x):
    if x['STADE'] != 'Exp�dition':
        for key in distances:
            if x['LIBELLE'].lower().find(key) > -1:
                return distances[key]
    return np.nan  
```

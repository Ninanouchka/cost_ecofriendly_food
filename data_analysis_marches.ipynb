```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_full = pd.read_csv('data_rnm_marches/cotations-RNM-54819.csv', sep=';', header=1)
df_full = df_full.append(pd.read_csv('data_rnm_marches/cotations-RNM-54856.csv', sep=';', header=1))
df_full = df_full.append(pd.read_csv('data_rnm_marches/cotations-RNM-37581.csv', sep=';', header=1))
df_full = df_full.append(pd.read_csv('data_rnm_marches/cotations-RNM-37624.csv', sep=';', header=1))
```

```python
df_full = df_full.dropna(subset=['Libellé'])
```

```python
distances = {'espagne': 1056, 'maroc': 1800, 'pays-bas': 430, 'ghana': 4815, 'martinique': 6930, 'cameroun': 5060, 'portugal': 1454, 'belgique': 265, 'jordanie': 3370, 'floride': 7200, 'isra�l': 3315, 'italie': 1105, 'chili': 11650, 'guadeloupe':6740, 'madagascar': 8600, 'egypte':3190, 'u.e.':1035, 'corse': 895, 'france': 0}
```

```python
def dist_origin(x):
    for key in distances:
        if x['Libellé'].lower().find(key) > -1:
            return distances[key]
    return np.nan 
```

```python
df_full['distance'] = df_full.apply(dist_origin, axis=1)
df_full['Moyen'] = df_full['Moyen'].str.replace(',', '.').astype(float)
```

```python
df_full
```

```python
df_full.groupby('Date').mean()['distance'].plot()
plt.ylabel('Distance')
plt.title('Variation of importation distance in 2019')
```

```python
df_full.drop(df_full[df_full['Unité prix'].str.find('le kg') == -1].index).groupby('Date').mean()['Moyen'].plot()
plt.ylabel('Price (€)')
plt.title('Variation of price in 2019')
```
